```python
class Casa:
  """
  Construir el constructor
  argumentos:
    color(str): color de la casa
    numero_habitaciones(int): el nmero de habitaciones de la casa
    numero_baos(int): el nmero de baos de la casas
    estacionamiento(bool): si tiene estacionamiento o no
    tiene_patio(bool): si tiene patio o no
  """
  def __init__(self, color, numero_habitaciones, numero_baos, estacionamiento, tiene_patio):
    self.color = color
    self.numero_habitaciones = numero_habitaciones
    self.numero_baos = numero_baos
    self.estacionamiento = estacionamiento
    self.tiene_patio = tiene_patio

  def descripcion(self):
    #Describir la casa
    descripcion=f"La casa es de color {self.color}, tiene {self.numero_habitaciones} habitaciones, {self.numero_baos} baos"
    if self.estacionamiento:
      descripcion += " y tiene estacionamiento."

    if self.tiene_patio:
      descripcion += " Adems tiene patio"
    return descripcion

  def pintar(self, nuevo_color):
    self.color=nuevo_color
    print(f"La casa ha sido pintada de color {self.color}")
```

```python
mi_casa=Casa("Negro",3, 2, True, True)
type(mi_casa)
```

```python
print(mi_casa.descripcion())
```

```python
mi_casa.tiene_patio=False
print(mi_casa.descripcion())
```

```python
mi_casa.pintar("negro")
```

```python
class Floristeria(Casa):
  """
  Constructor de la floristera
  Argumentos:
    color(str): color de la floristera
    numero_ventanas(int): el nmero de habitaciones de la floristera
    numero_puertas(int): el nmero de puertas de la floristera
    numero_baos(int): el nmero de baos de la floristera
    tipos_flores(list): Lista de flores
    nombre_floristera: nombre de la floristera
    wifi(bool): Si tiene acceso a internet o no
    numero_habitaciones(int): el nmero de habitaciones de la casa
    estacionamiento(bool): si tiene estacionamiento o no
    tiene_patio(bool): si tiene patio o no
  """
  def __init__(self, color, numero_habitaciones, numero_baos, estacionamiento, tiene_patio, numero_ventanas, numero_puertas, nombre_floristera, wifi, tipos_flores):
    super().__init__(color, numero_habitaciones, numero_baos, False, False)
    self.numero_ventanas = numero_ventanas
    self.numero_puertas = numero_puertas
    self.nombre_floristera = nombre_floristera
    self.wifi = wifi
    self.tipos_flores = tipos_flores
```

```python
mi_floristera=Floristeria("Negro", 2, 1, False, False, 1, 2, "La vie en Fleurs",True,["Rosas", "Girasoles", "Amapolas", "Claveles", "Lirios"])
mi_floristera.descripcion()
```
